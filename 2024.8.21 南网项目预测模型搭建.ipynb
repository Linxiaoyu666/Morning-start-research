新建一个预测模型用于南网数据

```python
# 新建筑的冷负荷数据
new_building_data = pd.read_csv('NanWang_sum_Aug.csv')
x = new_building_data.drop(columns=['Unnamed: 0','Cooling load kW','day','date2','RH','day_of_week'],axis = 1)
y = new_building_data['Cooling load kW']
from sklearn.preprocessing import StandardScaler
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.1, random_state=2024)

#   数据标准归一化
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_train
```

```python
## 带调参的lightGBM模型

def LGB_CV(
          max_depth,
          num_leaves,
          min_data_in_leaf,
          feature_fraction,
          bagging_fraction,
          lambda_l1
         ):
    
    # Convert X_train and Y_train to pandas DataFrame or Series
    X_train_df = pd.DataFrame(X_train)
    Y_train_series = pd.Series(Y_train)
    
    folds = KFold(n_splits=5, shuffle=True, random_state=15)
    oof = np.zeros(X_train.shape[0])

#     for fold_, (trn_idx, val_idx) in enumerate(folds.split(X_train, Y_train)):
#         print("fold n°{}".format(fold_))
#         trn_data = lgb.Dataset(X_train[trn_idx],
#                                label=Y_train[trn_idx],
#                                )
#         val_data = lgb.Dataset(X_train[val_idx],
#                                label=Y_train[val_idx],
#                                )
        
    for fold_, (trn_idx, val_idx) in enumerate(folds.split(X_train, Y_train)):
        print("fold n°{}".format(fold_))
        trn_data = lgb.Dataset(X_train_df.iloc[trn_idx],  # Using .iloc for indexing
                            label=Y_train_series.iloc[trn_idx])  # Using .iloc for indexing
        val_data = lgb.Dataset(X_train_df.iloc[val_idx],  # Using .iloc for indexing
                            label=Y_train_series.iloc[val_idx])  # Using .iloc for indexing
    
        param = {
            'num_leaves': int(num_leaves),
            'min_data_in_leaf': int(min_data_in_leaf), 
            'objective':'regression',
            'max_depth': int(max_depth),
            'learning_rate': 0.01,
            "boosting": "gbdt",
            "feature_fraction": feature_fraction,
            "bagging_freq": 1,
            "bagging_fraction": bagging_fraction ,
            "bagging_seed": 11,
            "metric": 'rmse',
            "lambda_l1": lambda_l1,
            "verbosity": -1
        }
    
        clf = lgb.train(param,
                        trn_data,
                        5000,
                        valid_sets = [trn_data, val_data],
                        verbose_eval=500,
                        early_stopping_rounds = 200)
        
        oof[val_idx] = clf.predict(X_train[val_idx],
                                   num_iteration=clf.best_iteration)
        
        del clf, trn_idx, val_idx
        
    return metrics.r2_score(oof, Y_train)
```

```python
from sklearn.datasets import load_boston
from bayes_opt import BayesianOptimization
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
LGB_BO = BayesianOptimization(LGB_CV, {
    'max_depth': (4, 20),
    'num_leaves': (5, 130),
    'min_data_in_leaf': (5, 30),
    'feature_fraction': (0.7, 1.0),
    'bagging_fraction': (0.7, 1.0),
    'lambda_l1': (0, 6)
    })
LGB_BO.maximize(init_points=100,n_iter=100)
```

```python
# 创建LightGBM数据集时设置特征预过滤参数并保持原始数据
train_data = lgb.Dataset(X_train, label=Y_train, free_raw_data=False)
valid_data = lgb.Dataset(X_test, label=Y_test, reference=train_data, free_raw_data=False)

# 使用初始模型继续训练，并添加验证集和早停功能
newpre_models = lgb.train(params, train_data, valid_sets=[valid_data],
                         num_boost_round=9000,
                         early_stopping_rounds=9000, verbose_eval= True)
```
